```python
from sklearn.feature_extraction.text import CountVectorizer

def count_vectorizer( data, max_features=5000):
    cv = CountVectorizer(max_features=max_features, stop_words="english")

    vector = cv.fit_transform(data["combined_features"].values.astype("U")).toarray()

    return vector
```

```python
vector = count_vectorizer(movie)
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity


def get_cos_sim(vector):
    cos_sim = cosine_similarity(vector)

    return cos_sim
```

```python
def top_5_movies(data, movie_title):
    vector = count_vectorizer(data)
    similarity = get_cos_sim(vector)

    idx = data[data["title"] == movie_title].index[0]

    distance = sorted(list(enumerate(similarity[idx])), reverse=True, key = lambda vector: vector[1])

    recommendation_list = []

    for i in distance[0:5]:
        recommendation_list.append(data.iloc[i[0]].title)

    return recommendation_list
```

```python
rec_list = top_5_movies(movie, "Avatar")

for i in rec_list:
    print(i)
```
